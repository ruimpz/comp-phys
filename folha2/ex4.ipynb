```python
from vpython import sphere, vector, color
import numpy as np
```

```python
L = 5
for i in range(L):
    for j in range(L):
        for k in range(L):
            if (i + j + k) % 2 == 0:
                sphere(pos = vector(i, j , k), radius = 0.3)
            else:
                sphere(pos = vector(i, j , k), color = color.blue, radius = 0.3)
```

```python
L = 5
for i in range(L):
    for j in range(L):
        for k in range(L):
            if (i + j + k) % 2 == 0:
                sphere(pos = vector(i, j , k), radius = 0.3)
            elif i % 2 != 0 or j % 2 != 0:
                sphere(pos = vector(i, j , k), color = color.blue, radius = 0.3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x, y = np.loadtxt("sunspots.txt", unpack = True)
plt.plot(x, y)
```

```python
plt.plot(x, y)

r = 5
z = .0 * x
for i in range(len(x)-r):
    z[i+r]=np.sum(y[i:i+r])
z /= 2*r
plt.xlim(5, 1000)
plt.plot(x, z)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

A equação de calor

$$\frac{\partial T}{\partial t}=D \frac{\partial^{2} T}{\partial x^{2}}$$

onde temos a condição 

$$T_{0}(t)=A+B \sin \frac{2 \pi t}{\tau}$$

```python
def T0(t):
    return 10 + 20 * np.sin(2 * np.pi * t / 365)
```

```python
D = .1
ti, tf = 0, 365 * 10
xi, xf = 0, 20
N_x, N_t = 200, 100000
a = (xf - xi) / N_x
h = (tf - ti) / N_t
ts = np.linspace(ti, tf, N_t)
xs = np.linspace(xi, xf, N_x)
T = np.ones((N_x, N_t), float) * 10
T[-1,:] = T0(ts)
T[0, :] = 11.
for i in range(1, N_t):
    T[1:-1, i] = T[1:-1, i-1] + D*h/a**2 * (T[2:,i-1] + T[:-2,i-1] - 2*T[1:-1,i-1])
plt.pcolormesh(T)
```

```python
t1 = int(tf / h * 9 / 10)
delta_t = int(t1 / 9 / 4)
for k in range(4):
    plt.plot(T[:, t1 + k*delta_t])
plt.show()
```

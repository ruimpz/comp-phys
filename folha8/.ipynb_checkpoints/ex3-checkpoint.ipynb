```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
a = .01
n = 101
V = 1
delta = 1e-6
A1 = np.zeros((n, n), float)
A2 = np.zeros((n, n), float)
rho = np.zeros((n, n), float)
A1[20:80, 20], A1[20:80, 80] = np.ones(60) * V, np.ones(60) * -V
error = 1.
while error > delta:
    error = 0.
    A2[1:-1, 1:-1] = 1/4 * (A1[:-2,1:-1] + A1[2:,1:-1] + A1[1:-1,2:] + A1[1:-1,:-2])
    A2[20:80, 20], A2[20:80, 80] = np.ones(60) * V, np.ones(60) * -V
    error = np.max(np.absolute(A2 - A1))
    A1 = np.copy(A2)
plt.pcolormesh(A2, cmap="GnBu")
```

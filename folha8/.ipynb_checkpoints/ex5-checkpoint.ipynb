```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
%matplotlib inline
```

Definimos

$$ \frac{\partial \phi}{\partial t}=v^{2} \frac{\partial^{2} \psi}{\partial x^{2}}$$

$$ \frac{\partial \psi}{\partial t}=\phi$$

Sendo as diferen√ßas finitas

$$ \phi_{i}(t+h)= \phi_{i}(t) + \frac{v^{2} h}{a^{2}}\left[ \psi_{i+1}(t)+\psi_{i-1}(t)-2 \psi_{i}(t)\right] $$

$$ \psi_{i}(t+ h) = \psi_{i}(t) + h \phi_{i}(t) $$


```python
def fcts_wave(psi0, phi0, v, xi, xf, ti, tf, N_a, N_h):
    a = (xf - xi)/N_a
    h = (tf - ti)/N_h
    xs = np.linspace(xi, xf, N_a)
    ts = np.linspace(ti, tf, N_h)
    Psi = np.zeros((N_a, N_h), float)
    Phi = np.zeros((N_a, N_h), float)
    Psi[:, 0] , Phi[:, 0] = psi0, phi0
    for i in range(1, N_h):
        Phi[1:-1, i] = Phi[1:-1, i-1] + (v/a)**2 * h * (Psi[:-2, i-1] + Psi[2:, i-1] - 2*Psi[1:-1, i-1])
        Psi[1:-1, i] = Psi[1:-1, i-1] + h * Phi[1:-1, i-1]
    return Psi, xs, ts
```

```python
xi, xf, ti, tf, v, N_a, N_h = 0, 1, 0, 2, 1, 100, 10000
Xs = np.linspace(xi, xf, N_a)
Ts = np.linspace(ti, tf, N_h)
psi0 = Xs * (1 - Xs) * np.exp(- (Xs-.1)**2 / 2 /.3)
phi0 = .0 * Xs
Psi, xs, ts = fcts_wave(psi0, phi0, v, xi, xf, ti, tf, N_a, N_h)
```

```python
fig = plt.figure()
ax = plt.axes(xlim=(xi, xf), ylim=(-1, 1))
line, = ax.plot([], [])

def init():
    line.set_data([], [])
    return line,
def animate(i):
    line.set_data(xs, Psi[:, i*100])
    return line,

anim = FuncAnimation(fig, animate, init_func=init,
                    frames=100, interval=30, blit=True)
```

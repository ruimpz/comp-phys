```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def fine_diffs(M, p, delta=1e-6):
    a = 1
    phi1 = np.zeros((M, M), float)
    phi2 = np.copy(phi1)
    phi2[1:-1, 1:-1] = 1/4*(phi1[:-2, 1:-1]+phi1[2:,1:-1]+phi1[1:-1,2:]+phi1[1:-1,:-2]+a**2*p[1:-1, 1:-1])
    error = np.absolute(phi2 - phi1).max()
    phi1, phi2 = phi2, phi1
    while error > delta:
        phi2[1:-1, 1:-1] =1/4*(phi1[:-2, 1:-1]+phi1[2:,1:-1]+phi1[1:-1,2:]+phi1[1:-1,:-2]+a**2*p[1:-1, 1:-1])
        error = np.absolute(phi2 - phi1).max()
        phi1, phi2 = phi2, phi1
    return phi2
```

```python
p = np.zeros((100, 100), float)
p[20:40, 20:40] = 1
p[60:80, 60:80] = -1
```

```python
plt.pcolormesh(fine_diffs(100, p))
```

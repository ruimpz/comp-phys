```python
from math import sqrt

def roots1(a, b, c):
    D = sqrt(b*b - 4 * a * c)
    return ((-b + D) /2 / a), ((-b + D) /2 / a)

def roots2(a, b, c):
    D = sqrt(b*b - 4 * a * c)
    return (2 * c / ( -b + D), 2 * c / ( -b - D))
```

```python
a, b, c = .001, 1000, .001
print(roots1(a, b, c))
print(roots2(a, b, c))
```

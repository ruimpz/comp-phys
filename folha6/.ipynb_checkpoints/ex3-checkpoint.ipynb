```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def dft(y):
    N = len(y)
    n = np.arange(N)
    c = np.zeros(N//2 + 1, complex)
    for k in range(N//2 + 1):
        c[k] = y.dot(np.exp(-2j * np.pi * k * n / N))
    return c
```

```python
trumpet = np.loadtxt("trumpet.txt")
plt.plot(trumpet[:10000])
plt.show()
plt.plot(np.abs(np.fft.fft(trumpet))[:10000])
plt.show()
```

```python
piano = np.loadtxt("piano.txt")
plt.plot(piano)
plt.show()
plt.plot(np.abs(np.fft.fft(piano))[:10000])
plt.show()
```

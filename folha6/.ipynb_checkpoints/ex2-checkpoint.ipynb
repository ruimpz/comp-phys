```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def dft(y):
    N = len(y)
    n = np.arange(N)
    c = np.zeros(N//2 + 1, complex)
    for k in range(N//2 + 1):
        c[k] = y.dot(np.exp(-2j * np.pi * k * n / N))
    return c
```

```python
x, y = np.loadtxt("sunspots.txt", unpack=True)
plt.plot(y)
```

```python
coeffs = dft(y)
plt.plot( np.abs(coeffs[2:50]) ** 2 )
plt.show()
```

```python
print(np.argmax(coeffs[1:50]))
```

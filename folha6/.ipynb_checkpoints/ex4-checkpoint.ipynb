```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt("dow.txt")
plt.plot(data)
plt.show()
```

```python
c = np.fft.rfft(data)
N = len(c)
c[N//10 + 1:] = 0
y = np.fft.irfft(c)
plt.plot(data)
plt.plot(y)
plt.show()
```

```python
c = np.fft.rfft(data)
N = len(c)
c[N//50 + 1:] = 0
y = np.fft.irfft(c)
plt.plot(data)
plt.plot(y)
plt.show()
```

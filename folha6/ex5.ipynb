```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Pretendemos suavizar um sinal da forma

$$f(t)=\left\{\begin{array}{cc}{1} & {\text { se }\lfloor 2 t\rfloor \text { é par, }} \\ {- 1} & {\text { se }\lfloor 2 t\rfloor \text { é impar, }}\end{array}\right.$$

```python
def f(x):
    y = .0 * x
    for i in range(len(x)):
        if 2 * x[i] // 1 % 2 == 0:
            y[i] = 1
        else:
            y[i] = -1
    return y
```

```python
t = np.linspace(0, 1, 1000)
y = f(t)
plt.plot(t, y)
```

```python
c = np.fft.rfft(y[:-1])
c[11:] = 0
y2 = np.fft.irfft(c)
plt.plot(y2)
plt.plot(y[:-1])
plt.show()
```

Não só vemos a suavização da onda, como vemos artefactos que se devem à remoção de frequências do sinal cuja soma levaria à convergência na onda original.

```python
import numpy as np
```

```python
def f(x):
    return x**4 - 2 * x + 1

def F(x):
    return x**5 / 5 - x**2 + x
```

```python
def simpson(a, b, N):
    N = N // 2 * 2
    h = (b - a) / N
    x = np.linspace(a, b, N + 1)
    y = f(x)
    return (h / 3) * ( y[0] + y[-1] + 4*y[1:-1:2].sum() + 2*y[2:-1:2].sum() )
```

```python
print(simpson(0, 2, 10))
```

```python
def error(a, b, N):
    return abs((F(b) - F(a)) - simpson(a, b, N)) / (F(b) - F(a)) * 100
```

```python
print(error(0, 2, 10))
```

```python
print(error(0, 2, 100))
```

```python
print(error(0, 2, 1000))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def E(x):
    return np.exp(-x**2)
```

```python
def simpson(a, b, h):
    x = np.arange(a, b + h, h)
    y = E(x)
    return (h / 3) * ( y[0] + y[-1] + 4*y[1:-1:2].sum() + 2*y[2:-1:2].sum() )
```

```python
print(simpson(0, 3, .1))
```

```python
a, b, h = 0, 3, .1
x = np.arange(a, b + h, h)
def iterated_integral(x):
    y = .0 * x
    for i in range(1, len(x)):
        y[i] = simpson(x[0], x[i], h)
    return y
```

```python
plt.plot(x, iterated_integral(x))
```

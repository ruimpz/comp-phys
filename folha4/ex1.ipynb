```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x, y = np.loadtxt("velocities.txt", unpack=True)

def trapezoid(x, y, r):
    return .5 * (y[0] + y[-1]) + y[1:r].sum()
```

```python
print(trapezoid(x, y, 100))
```

```python
def get_dist(x, y):
    dist = .0 * x
    for i in range(1, len(x)):
        dist[i] = trapezoid(x, y, i)
    return dist
```

```python
plt.plot(x, y)
plt.plot(x, get_dist(x, y))
```

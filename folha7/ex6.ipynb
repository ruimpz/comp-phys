```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def rk4(f, a, b, N, r0, params = []):
    h = (b - a) / N
    ts = np.linspace(a, b, N)
    r = np.zeros((len(r0), N), float)
    r[:, 0] = r0
    for i, t in enumerate(ts[1:]):
        r_t = r[:,i]
        k1 = h * f(r_t, t, *params)
        k2 = h * f(r_t + .5 * k1, t + .5 * h, *params)
        k3 = h * f(r_t + .5 * k2, t + .5 * h, *params)
        k4 = h * f(r_t + k3, t + h, *params)
        r[:,i + 1] = r_t + (k1 + 2 * k2 + 2 * k3 + k4) / 6
    return ts, r
```

```python
def f(r, t, w, n):
    return np.array([r[1], -w**2 * r[0]**n])
```

```python
t, r = rk4(f, 0, 50, 1000, [1, 0], [1, 1])
plt.plot(t, r[0])
plt.show()
```

```python
t, r = rk4(f, 0, 50, 1000, [2, 0], [1, 1])
plt.plot(t, r[0])
plt.show()
```

```python
def g(r, t, u, w):
    x, y = r
    return np.array([y, u*(1-x**2) * y - w**2 * x])
```

```python
us = [1, 2, 4]
for u in us:
    t, r = rk4(g, 0, 20, 1000, [1, 0], [u, 1])
    plt.plot(t, r[0])
plt.show()
```
